## So what is Jupyter, really?

- Tools for interactive computing
  - documents
  - interfaces
    - terminal
    - qtconsole
    - notebook
  - introspection
  - debugging
- A message protocol for remote execution
  - language-agnostic
  - rich remote repl

The protocol is [publicly documented](http://ipython.org/ipython-doc/dev/development/messaging.html), and [language agnostic](https://jupyter.org/).
